```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# creating a file list
file_list = []

# traversing the root directory
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# saving the file path to a variable
image1_file_path = file_list[16] # use squirel2.jpg
```

```python
# loading image and storing it in a variable
image_squirel2 = cv2.imread(image1_file_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Gaussian blur to smooth the image and reduce noise.
blurred = cv2.GaussianBlur(image_squirel2, (5, 5), 1.4)
```

```python
# Apply the Canny Edge Detector.
# The first threshold is the low threshold, and the second is the high threshold.
edges = cv2.Canny(blurred, threshold1=50, threshold2=150)
```

```python
# Display the original image and the edge detected image side by side.
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image_squirel2, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```
