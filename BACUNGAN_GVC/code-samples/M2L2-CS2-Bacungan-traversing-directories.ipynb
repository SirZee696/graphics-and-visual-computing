# Traversing Directories

```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

pwd checks the working directory of your .ipynb file

```python
pwd
```

you can use os.walk(), os.listdir(), os.scandir() to search your directories

```python
for (root, dirs, files) in os.walk(os.getcwd()):
    for d in dirs:
        print(d)
    for f in files:
        print(f)
```

```python
os.listdir(os.getcwd())
```

```python
obj = os.scandir()
for entry in obj:
    if entry.is_dir() or entry.is_file():
        print(entry.name)
```

```python
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))
```

```python
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
image_file_path = file_list[5] # Select the desired image by using its index. Select the index of cameraman.jpg and put in a new variable
```

```python
image_file_path
```

```python
image = cv2.imread(image_file_path)
```

```python
# Opening image using cv2.imshow()
cv2.imshow('image',image)
cv2.waitKey();
```

```python
# Opening image using plt.imshow()
plt.figure()
plt.title('Cameraman')
plt.imshow(image)
```
