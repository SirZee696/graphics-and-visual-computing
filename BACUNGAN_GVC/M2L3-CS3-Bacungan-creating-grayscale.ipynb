# Creating Grayscale

```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creates a 100x256 black canvas
canvas_size = (100, 256)
gray_image = np.zeros(canvas_size)
```

```python
gray_image
```

```python
# creates a 100x256 8bit black canvas
gray_image = np.zeros(canvas_size, dtype = np.uint8)
```

```python
gray_image
```

```python
# displays the grayscale image in using plt
plt.imshow(gray_image)
```

```python
gray_image.shape
```

```python
# creating a gradient on the x axis but iterating first on the y axis:
for y in range(gray_image.shape[1]): # the y axis
    for x in range(gray_image.shape[0]): # the x axis
        gray_image[x, y] = y # this draws the value of the y-axis
```

```python
plt.imshow(gray_image)
```

```python
# getting the value at [0][50] as coordinates [y][x] or [h][w] in cv2
gray_image[0][101]
```

```python
plt.imshow(gray_image, plt.cm.gray)
```

```python
# inverting the gradient
for x in range(gray_image.shape[0]): # the x axis
    for y in range(gray_image.shape[1]): # the y axis
        gray_image[x, y] = x # draws the value of the x-axis
```

```python
plt.imshow(gray_image)
```

```python
plt.imshow(gray_image, plt.cm.gray)
```

To create a color image just specify the third dimension of the array (for example (100,256,3)) and rest is same.

```python
color_image_canvas = np.zeros((100, 256, 3), dtype = np.uint8) # creates a 100x256 8bit black canvas
plt.imshow(color_image_canvas)
```
