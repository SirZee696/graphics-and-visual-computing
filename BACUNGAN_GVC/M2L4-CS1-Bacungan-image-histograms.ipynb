# Image Histogram

```python
# Importing libraries
import os
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

displaying a list of image files

```python
# creating a file list
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
image_file_path_lotion = file_list[12] # getting index of lotion image
image_file_path_splash = file_list[14] # getting index of splash image
```

uses cv2.IMREAD_GRAYSCALE to convert the colored imgaes into grayscale

```python
image_lotion = cv2.imread(image_file_path_lotion, cv2.IMREAD_GRAYSCALE) # converted to grayscale
image_splash = cv2.imread(image_file_path_splash, cv2.IMREAD_GRAYSCALE) # converted to grayscale
```

```python
# check whether conversion is successful
image_lotion.shape
```

```python
plt.figure(figsize=(9,9)) # create a 9x9 figure

plt.subplot(2, 2, 1) # create a subtplot with sizes (1, 2) at subplot 1
plt.title("grayscale lotion image")
plt.imshow(image_lotion, plt.cm.gray) 

plt.subplot(2, 2, 2)
plt.title("grayscale splash image")
plt.imshow(image_splash, plt.cm.gray)
```

```python
# Getting the historgram of each image
hist_lotion = cv2.calcHist([image_lotion], [0], None, [256], [0, 256])
hist_splash = cv2.calcHist([image_splash], [0], None, [256], [0, 256])
```

```python
# Plot the histograms using plt.plot
# NOTE: these are not images there are plots thats why plt.plot is used
plt.figure(figsize=(9,9))

plt.subplot(2, 2, 1)
plt.title("lotion histograph")
plt.plot(hist_lotion)

plt.subplot(2, 2, 2)
plt.title("splash histograph")
plt.plot(hist_splash)
```
