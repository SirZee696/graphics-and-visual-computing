```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# creating a file list
file_list = []

# traversing the root directory
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# saving the file path to a variable
image1_file_path = file_list[16] # use squirel2.jpg
```

```python
# loading image and storing it in a variable
image_squirel2 = cv2.imread(image1_file_path, cv2.IMREAD_GRAYSCALE).astype('float64')
```

```python
# Compute the Sobel gradient in the x-direction
grad_x = cv2.Sobel(image_squirel2, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

# Compute the Sobel gradient in the y-direction
grad_y = cv2.Sobel(image_squirel2, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)
```

```python
# Calculate the gradient magnitude using the Euclidean norm
gradient_magnitude = np.sqrt(grad_x**2 + grad_y**2)

# Normalize the magnitude image to the range [0, 255]
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

# Optionally, calculate the gradient angle (in degrees)
gradient_angle = np.arctan2(grad_y, grad_x) * (180.0 / np.pi)
```

```python
# Plot the original image, the edge magnitude, and the edge direction
plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_squirel2, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title("Sobel Edge Magnitude")
plt.axis('off')

plt.tight_layout()
plt.show()
```
