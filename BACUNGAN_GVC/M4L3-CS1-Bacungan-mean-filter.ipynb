```python
# additional library from scikit-image
from skimage.util import random_noise # adds noise to our image
```

check root directory

```python
pwd
```

```python
# creating a file list
file_list = []

# traversing the root directory
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# saving the file path to a variable
image1_file_path = file_list[6] # use box.jpg
```

```python
# loading image and storing it in a variable
img = cv2.imread(image1_file_path)
```

```python
plt.imshow(img)
```

```python
# create a 3x3 kernel
kernel_size = (3, 3)

# apply a mean filter to our image using cv2.blur()
mean_filtered_img = cv2.blur(img, kernel_size)
```

```python
# Display the original and filtered images
cv2.imshow('Original Image', img)
cv2.imshow('Mean Filtered Image', mean_filtered_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mean filter
#add salt and pepper noise to the image
noise_amount = random_noise(img, mode="s&p",amount=0.3)

# convert the image into uint8
noise_img = np.array(255*noise_amount, dtype = 'uint8')

# apply cv2.blur to noise_img
mean_filtered_img_wnoise = cv2.blur(noise_img, kernel_size)
```

```python
# Display the original and filtered images
cv2.imshow('Original Image w/ Noise', noise_img)
cv2.imshow('Mean Filtered Image', mean_filtered_img_wnoise)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
