# Image Rotation

```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

creating a list of image files

```python
# creating a file list
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# use OpenCV.jpg
image_file_path = file_list[7] 
```

```python
# loading the image file
img = cv2.imread(image_file_path)
```

```python
# Getting the image dimensions
height, width = img.shape[:2]
```

```python
# defining angle of rotation
angle = 45
```

```python
# determining the center of the image
center = (width/2, height/2)
```

```python
# creating the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1.0)
```

```python
# apply the rotation matrix
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))
```

```python
# display the original figure and rotated figure
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Rotated Image (45 degrees)")
plt.imshow(rotated_image)
plt.axis("off")

plt.show()
```
