displaying a list of image files

```python
# creating a file list
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# loading the image, use the squirel.jpg image
image_file_path_squirel = file_list[9] 
squirel = cv2.imread(image_file_path_squirel)

# getting the color channel
squirel_blue_channel = squirel[:,:,0]
squirel_green_channel = squirel[:,:,1]
squirel_red_channel = squirel[:,:,2]

# calculating the histogram of each color channel
squirel_blue_channel_hist = cv2.calcHist([squirel], [0], None, [256], [0, 256])
squirel_green_channel_hist = cv2.calcHist([squirel], [1], None, [256], [0, 256])
squirel_red_channel_hist = cv2.calcHist([squirel], [2], None, [256], [0, 256])
```

```python
# Display the images and plots
plt.figure(figsize=(12,12))

plt.subplot(5, 3, 2)
plt.imshow(cv2.cvtColor(squirel, cv2.COLOR_RGB2BGR))

plt.subplot(5, 3, 4)
plt.title("red channel", fontsize=4)
plt.imshow(squirel_red_channel, plt.cm.gray)

plt.subplot(5, 3, 5)
plt.title("green channel", fontsize=4)
plt.imshow(squirel_green_channel, plt.cm.gray)

plt.subplot(5, 3, 6)
plt.title("blue channel", fontsize=4)
plt.imshow(squirel_blue_channel, plt.cm.gray)

plt.subplot(5, 3, 7)
plt.title("red histogram", fontsize=4)
plt.hist(squirel_red_channel.flatten(), 256, [0,256], color="red")
plt.show

plt.subplot(5, 3, 8)
plt.title("green histogram", fontsize=4)
plt.hist(squirel_green_channel.flatten(), 256, [0,256], color="green")
plt.show

plt.subplot(5, 3, 9)
plt.title("blue histogram", fontsize=4)
plt.hist(squirel_blue_channel.flatten(), 256, [0,256], color="blue")
plt.show

# Combining all the plots
plt.subplot(5, 1, (4,5))
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    hist = cv2.calcHist([squirel], [i], None, [256], [0, 256])
    plt.plot(hist, color = color)

plt.title("Image Histogram", fontsize=4)
plt.show()
```
