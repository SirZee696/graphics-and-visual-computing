```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv2
import os
```

```python
root_dir = os.getcwd() # getting the root directory of your notebook
file = "cameraman.jpg" # file name inside the root directory of your notebook
```

to check whether your directory or your file exist use os.path.exist

```python
file_dir = os.path.join(root_dir, file) # joining the root directory and the file name
file_dir
```

```python
os.path.exists(file_dir) # checking root directory and file name if they exists
```

Testing whether the directory exist or not

```python
false_dir = os.path.join(root_dir, "not_true.JPG") # creating a false directory
os.path.exists(false_dir) # checking if directory exists
```

To read an image, use the function cv2.imread(filename[, flags]) where filename is the full path of image and flags specifies the way image should be read (>0 for color, =0 for greyscale, and <0 for loading image as is (with alpha channel)).

```python
image = cv2.imread(file_dir) # reading the image fileb
```

use cv2.imshow() cv2.namedWindow() or plt.imshow() to display images

```python
# displaying the image using cv2.imshow uses a different window
cv2.imshow('image', image) # show the image
cv2.waitKey(); # waiting for the image to close, ALWAYS INCLUDE A cv2.watiKey()
               # when using cv2.imshow() or notebook will crash
```

using cv2.namedWindow()

```python
cv2.namedWindow('name of window', cv2.WINDOW_NORMAL)
cv2.imshow('name of window', image) # show the image
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# displaying the image using plt.imshow uses the notebook
```

```python
plt.figure(figsize=(2, 2)) # use figsize() to adjust the size of the image
plt.title('Cameraman')
plt.plot()
plt.imshow(image)
```

```python
plt.figure(figsize=(4, 4)) # use figsize() to adjust the size of the image
plt.title('Cameraman')
plt.plot()
plt.imshow(image)
```

Use makedirs() for creating folders

The code below creates a resouce folder named resource and creates a path by joining the root_dir and resource_folder

```python
# creating a resource directory
resource_folder = 'resources' # the name of the resource folder
resource_dir = os.path.join(root_dir, resource_folder) # joining the root_dir and resouce_folder
```

```python
resource_dir
```

This creates a file path by joining the resouce_dir and file name

```python
new_file = os.path.join(resource_dir, file); # joining the resource_dir and file name
```

```python
new_file
```

Check whether path already exist

```python
if not os.path.exists(resource_dir): # use this if file already exist, this will create an file already exists error
    os.makedirs(resource_dir) # making the directory
```
