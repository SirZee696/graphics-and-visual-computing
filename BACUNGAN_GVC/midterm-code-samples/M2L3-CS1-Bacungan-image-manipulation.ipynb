you can use os.walk(), os.listdir(), os.scandir() to search your directories

```python
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
image_file_path = file_list[5] # Select the desired image by using its index. Select the index of cameraman.jpg and put in a new variable
```

```python
# IMAGES OPENNED IN plt has the following color format RGB
# image[:, :, 0] -> RED CHANNEL when opening with plt
# image[:, :, 1] -> BLUE CHANNEL when opening with plt
# image[:, :, 2] -> GREEN CHANNEL when opening with plt

# BUT WHEN OPENING WITH cv2 color format is BGR
# image[:, :, 0] -> BLUE CHANNEL when opening with cv2
# image[:, :, 1] -> GREEN CHANNEL when opening with cv2
# image[:, :, 2] -> RED CHANNEL when opening with cv2

image = cv2.imread(image_file_path) # reload image
image[:,:,1] = 0 # GREEN CHANNEL for cv2.imshow
image[:,:,2] = 0 # BLUE CHANNEL for cv2.imshow
red_image_pyplot = image
```

```python
image = cv2.imread(image_file_path) # reload image
image[:,:,0] = 0 # RED CHANNEL for cv2.imshow
image[:,:,1] = 0 # GREEN CHANNEL for cv2.imshow
blue_image_pyplot = image
```

```python
image = cv2.imread(image_file_path) # reload image
image[:,:,0] = 0 # RED CHANNEL for cv2.imshow
image[:,:,2] = 0 # BLUE CHANNEL for cv2.imshow
green_image_pyplot = image
```

```python
image = cv2.imread(image_file_path) # reload image
```

```python
# USING plt to display images display the RGB format
plt.figure(figsize=(9,9))

plt.subplot(2, 3, 1)
plt.title("red")
plt.imshow(red_image_pyplot)

plt.subplot(2, 3, 2)
plt.title("green")
plt.imshow(green_image_pyplot)

plt.subplot(2, 3, 3)
plt.title("blue")
plt.imshow(blue_image_pyplot)

plt.subplot(2, 3, 5)
plt.title("image in RGB")
plt.imshow(image)
```
