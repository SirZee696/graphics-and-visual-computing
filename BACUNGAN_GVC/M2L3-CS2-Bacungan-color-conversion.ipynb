# Color Conversion

```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

pwd checks the working directory of your .ipynb file

```python
pwd
```

you can use os.walk(), os.listdir(), os.scandir() to search your directories

```python
file_list = []
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# this creates a counter for array indexing
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
image_file_path = file_list[15]
```

```python
image_squirel = cv2.imread(image_file_path)
```

```python
# using cv2.imshow to open the squirel image 
# results in displaying the image in RGB format 
cv2.imshow('squirel', image_squirel)
cv2.waitKey();
```

```python
# using plt.imshow to open the squirel image 
# results in displaying the image in BGR format 
plt.figure(figsize=(4,4))
plt.imshow(image_squirel) # plt imshow shows the BGR color model
```

```python
image_squirel.shape
```

```python
# To display the squirel in RGB format using the plt.imshow, convert the the image into BGR
# by using the cv2.cvtColor(image, cv2.COLOR_RGB2BGR) function
image_squirel_BGR = cv2.cvtColor(image_squirel, cv2.COLOR_RGB2BGR) # convert squirel to RGB
```

```python
plt.figure(figsize=(4,4))
plt.imshow(image_squirel_BGR) # squirel image is in BGR format but display RGB format in plt.imshow
```

```python
# squirel image is in BGR format 
# and displays BGR format in cv2.imshow()
cv2.imshow('squirel', image_squirel_BGR)
cv2.waitKey();
```

```python
# convert squirel to GRAY
image_squirel_GRAY = cv2.cvtColor(image_squirel, cv2.COLOR_RGB2GRAY)
```

```python
image_squirel_GRAY.shape
```

```python
plt.figure(figsize=(3,3))
plt.title("Gray Scale in plt")
plt.imshow(image_squirel_GRAY) # squirel is in BGR
```

```python
cv2.imshow('Gray', image_squirel_GRAY) # cv2 displays the image in gray
cv2.waitKey();
```

```python
plt.figure(figsize=(3,3))
plt.title("To view Gray Scale in plt")
# plt.cm.gray this to view converted grayscale in plt
plt.imshow(image_squirel_GRAY, plt.cm.gray)
```

To see it in only 1 channel or gray you, need to convert it into grayscale
