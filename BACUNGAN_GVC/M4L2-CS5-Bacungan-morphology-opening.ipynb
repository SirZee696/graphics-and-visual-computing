```python
# Importing libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating a file list
file_list = []

# traversing the root directory
for (root, dirs, files) in os.walk(os.getcwd()):
    for f in files:
        if f.endswith(('jpg', 'jpeg', 'png')): # get all image files
            file_list.append(os.path.join(root, f))

# display all the image files with index number
counter = 0;
for file in file_list:
    print(counter,file)
    counter = counter + 1;
```

```python
# saving the file path to a variable
image1_file_path = file_list[7] # use car.png
```

```python
# Opening Implementation
# loading image in grayscale and storing it in a variable
img = cv2.imread(image1_file_path, cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow("image", img)
cv2.waitKey()
```

```python
# Optional: conert your image to a binary image using thresholding (if your image is not already binary)
# adjust the threshold value (127 here) according yo your needs
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Define a 3x3 rectangular structuring element (kernel)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))

# Appy morphological opening: erosion followed by dilation
eroded_binary = cv2.erode(binary, kernel, iterations=1)

# Apply dilation to the eroded image
eroded_dilated_image = cv2.dilate(eroded_binary, kernel, iterations=1)
```

```python
# Display the original images and 
# the subtracted image side by side
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(binary, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Image 3")
plt.imshow(eroded_dilated_image, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# using cv2.morphologyEx and cv2.MORPH_OPEN
opened_binary = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
```

```python
# Display the original images and 
# the subtracted image side by side
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(binary, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Image 3")
plt.imshow(opened_binary, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
