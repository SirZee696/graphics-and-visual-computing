```python
# canvas size
width, height = 20, 20

# Creata a new blank canvas (white background)
canvas = Image.new("RGB", (width, height), color="white")

# create a drawing context
draw = ImageDraw.Draw(canvas)

# draw a blue rectangle in the canvas
draw.rectangle([(2, 2), (15, 15)], fill="blue", outline="black");

# save the canvas as an image file (png)
canvas.save("blank_canvas_blue_rectangle.png")

# display the canvas
canvas.show()
```

```python
# choose pixel to inspect (e.g., inse the blue rectangle)
x, y = 3, 3 # the coordinamtes should be within the size of the canvas
pixel_value = canvas.getpixel((x,y)) # getting the pixel value

# display the pixel value 
print(f"Pixel at ({x}, {y}: {pixel_value})") # should be blue (0, 0, 255)
```

```python
# importing numpy libraries as np
import numpy as np

# convert the canvas to numpy array
pixel_matrix = np.array(canvas)

# print the RGB matrix
for row in pixel_matrix:
    print(row)
```
